```python
import pandas as pd

df_clean = pd.read_csv(
    "C:\ZEEL\projects\supply-chain-analytics\data\clean_supply_chain.csv"
)
```

```python
market_sales = (
    df_clean.groupby('Market')['Sales']
    .sum()
    .sort_values(ascending=False)
)

print(market_sales)
```

```python
category_sales = (
    df_clean.groupby('Category Name')['Sales']
    .sum()
    .sort_values(ascending=False)
)

print(category_sales)
```

```python
category_profit = (
    df_clean.groupby('Category Name')
    ['Order Profit Per Order']
    .sum()
    .sort_values(ascending=False)
)

print(category_profit)
```

```python
shipping_mode = (
    df_clean['Shipping Mode']
    .value_counts()
)

print(shipping_mode)
```

```python
late_pct = (
    df_clean['Late_delivery_risk']
    .mean()
    * 100
)

print(late_pct)
```

```python
country_sales = (
    df_clean.groupby('Order Country')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(country_sales)
```

```python
segment_sales = (
    df_clean.groupby('Customer Segment')
    ['Sales']
    .sum()
    .sort_values(ascending=False)
)

print(segment_sales)
```
